```python
# https://chatgpt.com/c/6729fa2b-04e4-800c-9c2e-e1dbb135e855

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df.dropna(how='any',inplace=True)
df.isnull().sum()
```

```python
# input data
x = df.iloc[:,1:-1].values # (Input data) In simpler terms: This line creates a NumPy array called x containing all rows and all columns except the first and last from the DataFrame df.
# output data
y = df.iloc[:,-1].values # (Output data) In simpler terms: This line creates a NumPy array called y containing all rows and only the last column from the DataFrame df.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import accuracy_score

def report(classifier):
  y_pred = classifier.predict(x_test)
  cm = confusion_matrix(y_test,y_pred)
  display = ConfusionMatrixDisplay(cm,display_labels=classifier.classes_)
  display.plot()
  print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
```

```python
# K-Nearest Neighbours Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
report(knn)
```

```python
# SVC

from sklearn.svm import SVC
svm = SVC(gamma='auto', random_state=10)
svm.fit(x_train,y_train)
```

```python
report(svm)
```
